```python
from pyarrow import fs, parquet as pq
import os
```

```python
bucket = os.environ['BUCKET']
```

```python
s3  = fs.S3FileSystem(region="us-west-2")
```

```python
songs_table = pq.read_table("s3://"+bucket+"/songs/songs.parquet") #pyarrow.lib.Table
```

```python
songs = songs_table.to_pandas()
```

```python
display(songs)
```

```python
artist_table = pq.read_table("s3://"+bucket+"/artists/artists.parquet") #pyarrow.lib.Table
```

```python
artists = artist_table.to_pandas()
```

```python
display(artists)
```
